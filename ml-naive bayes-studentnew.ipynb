```python
x=data.drop(['Class'],axis=1)
y=data['Class']
```

```python
a=x.select_dtypes(include='object')
```

```python
b=pd.get_dummies(a)
```

```python
from sklearn.feature_selection import chi2
```

```python
chi=chi2
```

```python
chi_scores=chi(b,y)
```

```python
chi_scores
```

```python
p=pd.Series(chi_scores[1],index=b.columns)
p
```

```python
p.sort_values(ascending=False, inplace=True)
p
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,20))
p.plot.bar()
```

```python
b.columns
```

```python
f=[]
for i,j in enumerate(p):
    if j<0.01:
        f.append(i)
```

```python
f
```

```python
z=p[f].index
```

```python
q=b[z]
```

```python
cont_var=x.select_dtypes(exclude='object')
```

```python
cont_var
```

```python
from scipy.stats import pearsonr
```

```python
y.value_counts()
```

```python
data['Class']=data['Class'].map({'M':1,'H':2,'L':3}).astype(int)
```

```python
corr=[]
for i in cont_var:
    m=pearsonr(cont_var[i], data['Class'])
    corr.append(m)
corr
```

```python
import numpy as np
x1=np.array(corr)
p1=pd.Series(x1[:,1],index=cont_var.columns)
p1
```

```python
p1.plot.bar()
```

```python
e=[]
for i, j in enumerate(p1):
    if j<0.01:
        e.append(i)
```

```python
n=p1[e].index
r=cont_var[n]
```

```python
final=pd.concat([q,r],axis=1)
final
```

```python
final.dropna()
```

```python
final.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(final,y,test_size=0.3, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(ytest,ypred)
```

```python
confusion_matrix(ytest, ypred)
```

```python
print(classification_report(ytest,ypred))
```
