```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('D:\Python training\ex1data1.txt',header=None)
df.head()
```

```python
df=df.values
print('df in the form of matrix ',df)
```

```python
Y=df[:,1]
Y
#print(Y)
```

```python
m=np.shape(df)[0]
print ('number of training sample ',m)
```

```python
def hypothesis_linear(X,theta):
    return np.matmul(X,theta)
```

```python
X=np.ones([m,2])
print ("x is equal to ",X)
Y=np.ones([m,1])
print ("y is equal to",Y)
Y[:,0]=df[:,1]
print ('new y is ', Y)
X[:,1]=df[:,0] 
print ('new x is ',X)
theta=np.array([[1],[0.3]])
print ('theta ie equal to',theta)
Y_predicted=hypothesis_linear(X,theta)
print ('predicted values for given theta 1 and 0.3 is',Y_predicted)
```

```python
plt.scatter(df[:,0],df[:,1],color=['red'])

plt.plot(df[:,0],list(Y_predicted))

plt.legend(('actual data','Predicted_data'))
plt.show()
```

```python
def compute_cost(X,Y,theta):
    temp=(np.matmul(np.transpose(np.matmul(X,theta)-Y),np.matmul(X,theta)-Y))
    return (0.5/m*temp)
```

```python
alpha=0.02
number_of_iteration=1500
```

```python
def compute_gradient(X,Y,theta,alpha,number_of_iteration):
    J_values=np.zeros([number_of_iteration,1])
    for i in range(0,number_of_iteration):
        h=np.matmul(X,theta)-Y
        change=(1.0/m)*(np.matmul(np.transpose(X),h))
        print ('change',change)
        
        theta=theta-alpha*change;
        J_values=compute_cost(X,Y,theta)
        print ('j values',J_values)
    return theta   

final_theta=compute_gradient(X,Y,theta,alpha,number_of_iteration)
print ('final_theta value is ',final_theta)
```

```python
linear_model=np.matmul(X,final_theta)
plt.scatter(df[:,0],df[:,1],color=['red'])

plt.plot(df[:,0],list(linear_model))

plt.legend(('actual data','Predicted_data'))
plt.show()
    
```
