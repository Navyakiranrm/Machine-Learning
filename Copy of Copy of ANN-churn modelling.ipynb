```python
gender_dummy=pd.get_dummies(data['Gender'],drop_first=True)
gender_dummy.head()
```

```python
geography_dummy=pd.get_dummies(data['Geography'],drop_first=True)
geography_dummy.head()
```

```python
df=pd.concat([data,gender_dummy,geography_dummy],axis=1)
```

```python
df.head()
```

```python
df=df.drop(['Gender','Geography','RowNumber','CustomerId','Surname'],axis=1)
```

```python
df.head()
```

```python
x=df.drop(['Exited'],axis=1)
y=df['Exited']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python
ytrain.head()
```

```python
ytest.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
len(xtrain)
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import ReLU,LeakyReLU,PReLU,ELU
```

```python
ANN=Sequential()
```

```python
ANN.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
ANN.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
ANN.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
ANN.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
reg = ANN.fit(xtrain,ytrain,validation_split=0.0,batch_size=10,epochs=100)
```

```python
y_pred=ANN.predict(xtest)
y_pred=y_pred>0.5
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(ytest,y_pred)
```

```python
confusion_matrix(ytest,y_pred)
```

```python
print(classification_report(ytest,y_pred))
```
