```python
print("Before OverSampling, counts of label '1': {}".format(sum(ytrain == 1))) 
print("Before OverSampling, counts of label '0': {} \n".format(sum(ytrain == 0))) 

from imblearn.over_sampling import SMOTE
sm=SMOTE(sampling_strategy='minority')
xsmot,ysmot=sm.fit_sample(xtrain,ytrain)

print('After OverSampling, the shape of train_X: {}'.format(xsmot.shape)) 
print('After OverSampling, the shape of train_y: {} \n'.format(ysmot.shape)) 
  
print("After OverSampling, counts of label '1': {}".format(sum(ysmot == 1))) 
print("After OverSampling, counts of label '0': {}".format(sum(ysmot == 0))) 
```

```python
xsmot.shape
```

```python
import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import ReLU,LeakyReLU,ELU
```

```python
ann=Sequential()
```

```python
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=xsmot.shape[1]))
ann.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
ann.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
ann.compile(optimizer='adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
reg = ann.fit(xsmot,ysmot,validation_split=0.0,batch_size=10,epochs=100)
```

```python
ypred=ann.predict(xsmot)
ypred=ypred>0.5
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(ysmot,ypred)
```

```python
confusion_matrix(ysmot,ypred)
```

```python
print(classification_report(ysmot,ypred))
```
